```python
#prezicera nivelului apei in cm a Lacului Beleu
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel('beleu_regim_hidrologic.xlsx')
df
```

```python
# creem variabile pentru prezicerea in zile a viitorului
projection = 365
#creem coloana care prezicem
df['Prezicere'] = df[['Regimul hidrologic in cm']].shift(-projection)
#vedem rezultatul
df
```

```python
#creem data setul independent (x)
X = np.array(df[['Regimul hidrologic in cm']])
#stergem ultimele 365 rinduri
X = X[:-projection]
print(X)
```

```python
#creem un data set (y)
y = df['Prezicere'].values
y = y[:-projection]
print(y)
```

```python
# despartim datele in 85% antrenare si 15% antrenare
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= .365)
```

```python
#creem un model de antrenare
linReg = LinearRegression()
#antrenam modelul
linReg.fit(X_train, y_train)
```

```python
#testam modelul folosind scorul
linReg_confidence = linReg.score(X_test, y_test)
print('Linear regression Confidence:', linReg_confidence)
```

```python
#creem o variabila x_projection pentru ultimele 365 de rinduri din data setul original
x_projection = np.array(df[['Regimul hidrologic in cm']])[-projection:]
print(x_projection)
```

```python
# afisam prezicerea pentru pentru urm 365 zile
linReg_prediction = linReg.predict(x_projection)
print(linReg_prediction)
```
